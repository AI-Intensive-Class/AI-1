```python
import pandas as pd
df = pd.DataFrame(columns=['index', 'filename', 'label'])
idx=0

# using tqdm
from tqdm import tqdm
```

```python
batch_size = 24
torch.cuda.empty_cache()
dataset_len = len(shapenet_dataset)

for i in tqdm([1]):
  a,b,c = values[i]
  R, T = look_at_view_transform(a, b, c)
  cameras = OpenGLPerspectiveCameras(R=R, T=T, device=device)
  raster_settings = RasterizationSettings(image_size=512,cull_backfaces=True)
  lights = PointLights(
    device=device,
    location=[[0.0, 5.0, -10.0]],
    diffuse_color=((0, 0, 0),),
    specular_color=((0, 0, 0),),
    )
  print(idx)
  # per batch size
  # for j in tqdm(range(0,dataset_len,batch_size)):
  j=0
  loop=0
  with tqdm(total=dataset_len) as pbar:
   while j < dataset_len:
     # print(j,idx)
     max_len = min(j+batch_size,dataset_len)

     temp_idxs = list(range(j,max_len))
     shapenet_objects = []
     object_synset_ids = []
     object_model_ids = []

     for id in temp_idxs[:]:
       try:
         temp_object = shapenet_dataset[id]
         # if i==6 and os.path.exists(f"dataset6/{temp_object['model_id']}_6.png") :
         #   df.loc[idx] = [idx, f"{temp_object['model_id']}_6.png", temp_object["synset_id"]]
         #   df.to_csv("dataset67/dataset67.csv", index=False)
         #   idx+=1
         #   temp_idxs.remove(id)
         #   continue
         shapenet_objects.append(temp_object)
         object_synset_ids.append(temp_object["synset_id"])
         object_model_ids.append(temp_object["model_id"])
       except:
         temp_idxs.remove(id)
         continue

     try:
       imgs  = shapenet_dataset.render(
    idxs=temp_idxs,
    device=device,
    cameras=cameras,
    raster_settings=raster_settings,
    lights=lights,
    shader = HardFlatShader(device=device,cameras=cameras,lights=lights)
    )
     except Exception as e:
       print(f"Rendering failed at batch {j}, idx {idx}: {e}")
       batch_size //= 2
       batch_size = max(1,batch_size)
       print(f"Reducing batch size to {batch_size}")
       torch.cuda.empty_cache()
       loop=0
       continue


    # print(idx)
     for k in range(len(temp_idxs)):
       plt.imsave(f"dataset1/{object_model_ids[k]}_{i}.png",imgs[k,...,:3].cpu().numpy())
       df.loc[idx] = [idx, f"{object_model_ids[k]}_{i}.png", object_synset_ids[k]]
       df.to_csv("dataset1/dataset1.csv", index=False)
       idx+=1
      # print(f"{object_model_ids[k]}_{i}.png : {object_synset_ids[k]}")

     j+=batch_size
     pbar.update(batch_size)
    # # Free the GPU memory
     del imgs
     torch.cuda.empty_cache()
     gc.collect()

     loop+=1
     if loop > 20:
       batch_size *=2
       batch_size = min(batch_size,24)
       loop=0
       print(f"Increasing batch size to {batch_size}")

```
